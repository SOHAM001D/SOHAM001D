```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

print('Libraries imported succesfully')
```

```python
#Loading the dataset
DF = pd.read_csv(r'C:\Users\Soham001\real-estate-price-prediction\Dataset .csv')
display(DF)
```

```python
#Deleting the missing or Null values in cuisines column
new_cuisines = DF['Cuisines'].dropna()
display(DF)
```

```python
#Split and count Cuisines
Cuisines_count = DF['Cuisines'].str.split(',',).explode().value_counts()
top_three_Cuisines = Cuisines_count.head(3)
display(top_three_Cuisines)
```

```python
#Creating Bar chart for top 3 most common Cuisines
colors = ['blue','yellow','red']
plt.bar(top_three_Cuisines.index, top_three_Cuisines.values, color=colors)
plt.title('Top 3 most common Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.show()
```

```python
#Calculating Percentage of restaurants that serve each top Cuisines
total_rest = len(new_cuisines)
percentage_Cuisines = (top_three_Cuisines / total_rest) * 100
print(percentage_Cuisines)
```

```python
#Creating pie chart for percentage
plt.pie(top_three_Cuisines.values,labels=top_three_Cuisines.index, autopct='%1.1f%%')
plt.title('Piechart for Percentage of restaurants that serve each top Cuisines ')
plt.show()
```
